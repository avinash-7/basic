```python
import pandas as pd 
df=pd.read_csv('Data_1.csv') # column names can be changed in this way also 
print(df)

newdf=df.fillna(0) # replace the missing values by 0 
print('\n\n',newdf)

newdf2=df.fillna({'Age':0,'Annual Salary':1}) # replace the missing values by 0 in Age 
                                            # replace the missing values by 1 in Annual Salary column
print('\n\n',newdf2)
```

```python
# now we see that 0 and 1 doest make sense always because we see that age cant be 0 or salary can be 1
import pandas as pd 
df=pd.read_csv('Data_1.csv') # column names can be changed in this way also 
print(df)

newdf=df.fillna(method='ffill') # replace the missing values by upcoming value
print('\n\n',newdf)

newdf2=df.fillna(method='bfill') # replace the missing values by previous vaue
print('\n\n',newdf2)

# similarly we can add the value of previous column or upcoming column by using axis =columns
newdf3=df.fillna(method='bfill',axis=1) # replace the missing values by previous vaue
print('\n\n',newdf3)
```

# grouping 

```python
import pandas as pd
df=pd.read_csv('Data_1.csv') 
print(df)

print('\n\n\n  here we will see the datatypes of all the columns present in our data',df.dtypes)

gf=df.groupby('Location')
print('\n\n\n',gf,type(gf))
for k,v in gf: # where k will show our subgroup and v will show its data 
    print('------------------------------------------------')
    print(k)
    print('---------------------')
    print(v)
    
```
